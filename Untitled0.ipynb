```python
# Um teste em que o usuário digita manualmente a nota do aluno e o programa retorna se ele obteve conceito A ou não.
def conceitoa(nota):
    if nota >= 90:
        return 1
    else:
        return 0

nota = int(input("Digite sua nota: "))
resultado = conceitoa(nota)

print("Resultado da nota =", resultado)
```

```python
# Um teste automatizado que lê um conjunto de notas e executa a função para verificar os resultados.
def conceitoa(nota):
    if nota >= 90:
        return 1
    else:
        return 0
notas = [50, 100, 95, 73]

for nota in notas:
  resultado = conceitoa(nota)
  print("Nota", nota, "Resultado da nota =", resultado)
```

```python
#c) Uma versão aprimorada do teste do item (b), tornando-o mais robusto e confiável.

def conceitoa(nota):
    if nota >= 90:
        return 1
    else:
        return 0

notas = [50, 100, 95, 73, 15, 79, 99, 30]
conceito_esperado = [0, 1, 1, 0, 1, 0, 0, 1]

print("--- Iniciando Teste  ---")

for i in range(len(notas)):

    nota_atual = notas[i]
    resultado_real = conceitoa(nota_atual)
    resultado_que_eu_queria = conceito_esperado[i]

    # If e Else para retornar o Conceito A
    if resultado_real == resultado_que_eu_queria:
        print("✅ OK! Nota", nota_atual, "retornou", resultado_real, "como esperado.")
    else:
        print("❌ NAO E A NOTA! Nota", nota_atual, "deu", resultado_real, "mas eu esperava", resultado_que_eu_queria)
```
