```python
# Como elaborar uma função
# Exemplo: Crie uma rotina que testa a função soma.
# rotina soma que será testada
def soma(num1, num2):
  soma = num1 + num2
  return (soma)

# rotina que testa a rotina soma.
def testar_soma(num1, num2, resultado_esperado):
  resultado_soma=soma(num1,num2) # função que será testada
  if resultado_soma == resultado_esperado:
    print("Teste passou: {} + {} = {}".format(num1, num2, resultado_soma))
  else:
    print("Teste falhou: {} + {} = {} (esperado: {})".format(num1, num2, resultado_soma, resultado_esperado))


# programa principal que chama a função
# chama a função soma passando os números a serem somados e o resultado esperado
# 5 casos de teste
testar_soma(5,3,8) # caso de teste 1
testar_soma(5,7,12) # caso de teste 2
testar_soma(7,13,20) # caso de teste 3
testar_soma(15,7,22) # caso de teste 4
testar_soma(1,2,4) # caso de teste 5. esse caso de teste está errado
```

```python
# O teste vai solicitar as entradas para o testador (usuário)
# O código irá permitir a entrada de vários casos de teste. (laço)
# rotina soma que será testada
def soma(num1, num2):
  soma = num1 + num2
  return (soma)

# função para testar a rotina soma
def testar_soma(num1, num2, resultado_esperado):
  resultado_soma=soma(num1,num2) # função que será testada
  if resultado_soma == resultado_esperado:
    print("Teste passou: {} + {} = {}".format(num1, num2, resultado_soma))
  else:
    print("Teste falhou: {} + {} = {} (esperado: {})".format(num1, num2, resultado_soma, resultado_esperado))

# programa principal que chama a função
# Tester entra com os vários casos de teste
num1=1
while num1 != 0:
  num1= int(input("entre com o primeiro número da soma="))
  num2= int(input("entre com o segundo número da soma="))
  resultado= int(input("entre com o resultado esperado"))
  # chama a função teste enviando os parâmetros digitados
  testar_soma(num1,num2,resultado)
print ("fim do programa")
```

```python
'''
PROGRAMA 3 - Testa se a função add está correta com VÁRIAS ENTRADAS ALEATÓRIAS USANDO FUNÇâO RANDON
Quando NÃO conhecemos os testes, chamamos de TESTE CAIXA PRETA
'''
# define the function add (função que será testada)
def add(a, b):
    return a + b

# função test
def testsoma(a,b,resultado):
  soma=add(a,b)
  print("Resultado da soma ",a, "+", b, "=", soma, " O resultado esperado é: ", resultado)
  if resultado != soma:
    print ("função add está ERRADA.")
  return

# Testa se a função add está correta com VÁRIAS ENTRADAS ALEATÓRIAS. Teste de exaustão
# MASSA DE DADOS para Testar se a função add está correta com VÁRIAS ENTRADAS
# Exemplo de teste de CAIXA PRETA e pode ser usado para teste de carga.
import random
# Testa COM VÁRIAS ENTRADAS
for contador in range(10):
  a=random.randrange(1, 100)
  b=random.randrange(1, 100)
  resultado = a + b
  testsoma(a, b, resultado)
print ("fim de programa")

```

```python
#  PROGRAMA 4 - Testa se a função add está correta
# com VÁRIAS ENTRADAS COM USO DE ARQUIVO
# Criar um arquivo de teste, ler e testar cada registro

# define the function add (função que será testada)
def add(num1, num2):
    return num1 + num2

# define a função test para testar várias entradas
def testsoma(num1,num2,resultado_esperado):
  soma=add(num1,num2)
  print (" ")
  print("Resultado da soma ",num1, "+", num2, "=", soma, " O resultado esperado é: ", resultado_esperado)
  if resultado_esperado == soma:
    print ("função add está correta.")
  else:
    print("\033[31m***** função add está ERRADA! *****\033[0m")
  return

# define a função que cria o arquivo com a MASSA DE DADOS para Testar
# se a função add está correta com VÁRIAS ENTRADAS VIA ARQUIVO.
# gravar no arquivo testecalculadora.txt registros no formato de 2 números que serão somados
# e o resultado da soma.
with open("testecalculadora.txt", "w") as arquivo:
  arquivo.write("1, 2, 3\n")
  arquivo.write("2, 5, 7\n")
  arquivo.write("5, 5, 10\n")
  arquivo.write("7, 9, 16\n")
  arquivo.write("12345678, 987, 12346665\n")
  arquivo.write("7, 10, 17\n")
  # na carga abaixo há um erro nos dados do teste
  arquivo.write("7, 9, 10")
arquivo.close()

# abrir e ler o arquivo testecalculadora.tx até o final e
# mostrar cada registro lido.
with open("testecalculadora.txt", "r") as arquivo:
  print("Mostrando todas os registros do arquivo testecalculadora")
  for registro in arquivo:
    print(registro)

# Faca um programa que leia o arquivo registros e chame a função def testsoma(a,b,resultado) para cada registro lido
# Extract values from each line
# Convert values to appropriate types
# Call the function with extracted values
print (" ")
print("Mostrando os resultados do teste")
with open("testecalculadora.txt", "r") as arquivo:
  for registro in arquivo:
    num1, num2, resultado_esperado = registro.strip().split(",")
    num1 = int(num1)
    num2 = int(num2)
    resultado_esperado = int(resultado_esperado)
    testsoma(num1, num2, resultado_esperado)
```
